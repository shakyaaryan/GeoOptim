## Road Functional Grouping

The original Overture road classifications were consolidated into three broader functional groups to simplify feature engineering and reduce the number of road categories used in the machine learning model. Grouping similar road types improves model interpretability while preserving the primary functional characteristics of the road network.

| Functional Group | Original Road Classes | Description |
|------------------|----------------------|-------------|
| **Arterial_Vehicular** | `motorway`, `trunk`, `primary`, `secondary` | Major roads designed for high-capacity vehicular movement, connecting cities, districts, and major urban areas. These roads typically carry the highest traffic volumes. |
| **Local_Commercial** | `tertiary`, `residential`, `living_street`, `service`, `unclassified` | Roads that provide access to residential neighborhoods, commercial areas, and local destinations. They primarily support local traffic and property access. |
| **Pedestrian_Only** | `pedestrian`, `footway`, `path`, `steps`, `cycleway` | Infrastructure intended primarily for pedestrians and cyclists, including sidewalks, walking paths, stairways, and dedicated cycling routes. These roads generally have limited or no access for motor vehicles. |

### Purpose of Grouping

The road classes were aggregated to:

- Reduce category sparsity in the dataset.
- Simplify spatial feature engineering.
- Capture the functional role of roads instead of fine-grained classifications.
- Improve the robustness and interpretability of the machine learning model.

```python
import geopandas as gpd

# 1. Define the functional group map
road_group_map = {
    # High-speed vehicular arteries
    'motorway': 'Arterial_Vehicular', 'trunk': 'Arterial_Vehicular', 
    'primary': 'Arterial_Vehicular', 'secondary': 'Arterial_Vehicular',
    
    # Neighborhood commercial/residential links
    'tertiary': 'Local_Commercial', 'residential': 'Local_Commercial', 
    'living_street': 'Local_Commercial', 'service': 'Local_Commercial', 
    'unclassified': 'Local_Commercial',
    
    # Pedestrian infrastructure
    'pedestrian': 'Pedestrian_Only', 'footway': 'Pedestrian_Only', 
    'path': 'Pedestrian_Only', 'steps': 'Pedestrian_Only', 
    'cycleway': 'Pedestrian_Only'
}

print("Loading light road layer...")
roads_gdf = gpd.read_file(r"C:\Users\anils\Desktop\7th SEM\GIS\GeoOptim\Data\Overture\light_roads.geojson")

# 2. Map the raw Overture classes to your clean functional layers
roads_gdf['functional_class'] = roads_gdf['class'].map(road_group_map)

# 3. Drop unmapped classes (track, bridleway, unknown) which act as model noise
roads_gdf.dropna(subset=['functional_class'], inplace=True)

# 4. Save only the essential columns for your shop proximity matrix loops
final_roads = roads_gdf[['id', 'functional_class', 'geometry']]

print("\n--- Cleaned Kathmandu Functional Road Segment Distribution ---")
print(final_roads['functional_class'].value_counts())

# Save the master road network
final_roads.to_file(r"C:\Users\anils\Desktop\7th SEM\GIS\GeoOptim\Data\Overture\functional_roads.geojson", driver="GeoJSON")
print("\nFile 'kathmandu_functional_roads.geojson' successfully structured and ready!")
```

## Raster Clipping to the Study Area

The original raster datasets (e.g., **WorldPop Population** and **Copernicus DEM**) cover a much larger geographic extent than the study area. To reduce storage requirements and improve processing efficiency, each raster was clipped to the Kathmandu study area's bounding box.

The clipping operation was performed using the `rasterio.mask` module, where a rectangular bounding box was defined using its minimum and maximum longitude and latitude coordinates.

### Bounding Box Coordinates

| Parameter | Value |
|----------|-------|
| Minimum Longitude | 85.20° E |
| Minimum Latitude | 27.60° N |
| Maximum Longitude | 85.50° E |
| Maximum Latitude | 27.75° N |

The bounding box was used to mask the original raster, retaining only the pixels that intersect the study area. The output raster preserves the original spatial resolution, coordinate reference system (CRS), and pixel values while reducing the file size and computational overhead for subsequent spatial analysis and feature engineering.

```python
import rasterio
from rasterio.mask import mask
from shapely.geometry import box

# 1. Define the Kathmandu bounding box
bbox = box(85.20, 27.60, 85.50, 27.75)
input_raster = RAW_DATA/"npl_pop_2025_CN_100m_R2024B_v1.tif"

# 2. Open the source and crop
with rasterio.open(input_raster) as src:
    out_image, out_transform = mask(src, [bbox], crop=True)
    out_meta = src.meta.copy()

# 3. Update the metadata for the new, smaller file
out_meta.update({
    "driver": "GTiff",
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})

# 4. Save the cropped raster
output_path = DATA_DIR/"kathmandu_population_density.tif"
with rasterio.open(output_path, "w", **out_meta) as dest:
    dest.write(out_image)

print(f"Successfully saved cropped population density to: {output_path}")
```
