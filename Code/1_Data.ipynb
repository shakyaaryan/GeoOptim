```python
import rasterio
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

# 1. Load your saved Kathmandu population file
with rasterio.open(DATA_DIR/"kathmandu_population_density.tif") as src:
    pop_data = src.read(1)  # Read the first band
    # Mask out no-data values if they exist (usually < 0)
    pop_data = pop_data.astype(float)
    pop_data[pop_data < 0] = 0

# 2. Plotting with a large figure and logarithmic scaling
fig, ax = plt.subplots(figsize=(12, 10))

# Using 'magma' colormap for clear visualization of subtle density gradients
img = ax.imshow(
    pop_data, 
    cmap='magma', 
    norm=LogNorm(vmin=1, vmax=pop_data.max()),
    extent=[85.20, 85.50, 27.60, 27.75]
)

# 3. Aesthetics
cbar = fig.colorbar(img, ax=ax, fraction=0.035, pad=0.04)
cbar.set_label('Population per 100m Grid Cell (Log Scale)')
ax.set_title('Kathmandu Population Density - Saved Layer', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
